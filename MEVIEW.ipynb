```python
import cv2 
import os 
video = cv2.VideoCapture(r'/content/drive/My Drive/microexpression_meview/dataset/me-cuts/cuts/sub01-1.mp4') 
try:  
	if not os.path.exists('/content/drive/My Drive/microexpression_meview/dataset/neutral'): 
		os.makedirs('/content/drive/My Drive/microexpression_meview/dataset/neutral') 
except OSError: 
	print ('Error') 
currentframe = 0
while(True): 
	ret,frame = video.read() 

	if ret: 
		name = '/content/drive/My Drive/microexpression_meview/dataset/neutral/sub01-1-' + str(currentframe) + '.jpg'
		print ('Captured...' + name) 
		cv2.imwrite(name, frame) 
		currentframe += 1
	else: 
		break
video.release() 
cv2.destroyAllWindows()
```
